```python
import pandas as pd
from sklearn.cluster import DBSCAN

books_df = pd.read_parquet('..\\data\\transformed\\books_description.parquet')
books_df
```

```python
books_df = books_df.reset_index()
books_df
```

```python
books_df['temp'] = books_df['id']
books_df['id'] = books_df['index']
books_df['vectors'] = books_df['temp']
#books_df.drop(columns={'temp', 'index'})
books_df
```

```python
books_df = books_df.drop(columns = {'temp', 'index'})
books_df
```

```python
metadata_df = pd.read_parquet('..\\data\\transformed\\books_metadata.parquet')
metadata_df
```

```python
metadata_df = metadata_df.reset_index()
metadata_df['id'] = metadata_df['index']
metadata_df = metadata_df.drop(columns = {'index'})
```

```python
merged = books_df.join(metadata_df, on='id', lsuffix='id')
```

```python
merged.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

X = merged[['id', 'vectors']]
clustering = DBSCAN(eps=3, min_samples=2).fit(X)
clustering.labels_
clustering
```
