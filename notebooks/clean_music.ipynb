```python
import pandas as pd
```

```python
metadata = pd.read_parquet("../data/raw/music_metadata.parquet")
metadata
```

```python
len(metadata['id']) == len(set(metadata['id']))
```

```python
set(metadata['year'])
```

```python
metadata = metadata[metadata['views'] > metadata['views'].quantile(0.6)]
```

```python
for i in range(1, 25):
    metadata.loc[metadata['year'] == i, 'year'] = int(f"200{i}") if i < 10 else int(f"20{i}")
```

```python
metadata = metadata[metadata['year'] > 1850]
```

```python
metadata
```

```python
import re

def split_features(s: str):
    s = re.sub(r'\\', '', s)
    words = re.findall(r'\"(.*?)\"', s)
    return words[:3] + [None] * (3 - len(words))

metadata[['feature_1', 'feature_2', 'feature_3']] = metadata['features'].apply(split_features).apply(pd.Series)
```

```python
metadata = metadata.drop(columns=["features"])
metadata
```

```python
metadata['total_features'] = metadata.apply(lambda row: sum(pd.notna(row[col]) for col in ['feature_1', 'feature_2', 'feature_3']), axis=1)
metadata
```

```python
metadata.to_parquet("../data/raw/music_metadata.parquet")
```
