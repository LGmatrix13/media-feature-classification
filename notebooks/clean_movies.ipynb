```python
import pandas as pd
```

```python
metadata = pd.read_parquet("../data/raw/movies_metadata.parquet")
metadata
```

```python
len(metadata)
```

```python
len(set(metadata['id']))
```

```python
metadata = metadata.drop_duplicates(subset=["id"], keep='first')
```

```python
len(metadata)
```

```python
metadata = metadata[metadata['title'] != 'Untitled']
metadata.head()
```

```python
metadata[['keyword_1', 'keyword_2', 'keyword_3', 'keyword_4', 'rest']] = metadata['keywords'].str.split(',', n=4, expand=True)
```

```python
metadata[['keyword_1', 'keyword_2', 'keyword_3', 'keyword_4']]
```

```python
metadata = metadata[metadata['status'] != 'Canceled']
```

```python
metadata[['genre_1', 'genre_2', 'genre_3', 'genre_4', 'genre_rest']] = metadata['genres'].str.split(',', n=4, expand=True)
```

```python
metadata.columns
```

```python
metadata[['production_company_1', 'production_company_2', 'production_company_3', 'production_company_4', 'production_company_rest']] = metadata['production_companies'].str.split(',', n=4, expand=True)
```

```python
metadata.columns
```

```python
metadata = metadata.drop(columns=["production_company_rest", "genre_rest", "rest", "spoken_languages", "production_companies", "genres", "poster_path", "original_title", "original_language", "imdb_id", "homepage", "backdrop_path"])
```

```python
metadata[['production_country_1', 'production_country_2', 'production_country_3', 'production_country_rest']] = metadata['production_countries'].str.split(',', n=3, expand=True)
```

```python
metadata = metadata.drop(columns=["production_country_rest", "keywords", "production_countries"])
```

```python
metadata[metadata['revenue'] == 0]
```

```python
metadata['total_keywords'] = metadata.apply(lambda row: sum(pd.notna(row[col]) for col in ['keyword_1', 'keyword_2', 'keyword_3', 'keyword_4']), axis=1)
metadata['total_genre'] = metadata.apply(lambda row: sum(pd.notna(row[col]) for col in ['genre_1', 'genre_2', 'genre_3', 'genre_4']), axis=1)
metadata['total_production_company'] = metadata.apply(lambda row: sum(pd.notna(row[col]) for col in ['production_company_1', 'production_company_2', 'production_company_3', 'production_company_4']), axis=1)
metadata['total_production_country'] = metadata.apply(lambda row: sum(pd.notna(row[col]) for col in ['production_country_1', 'production_country_2', 'production_country_3']), axis=1)
metadata
```

```python
metadata.to_parquet('../data/raw/movies_metadata.parquet')
```
